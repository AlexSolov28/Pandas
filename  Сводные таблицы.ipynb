```python
# Меняет тип даныых дат из object в datetime64[ns]
df_o['OrderDate'] = df_o['OrderDate'].astype('datetime64[ns]')
df_o['RequiredDate'] = df_o['RequiredDate'].astype('datetime64[ns]')
df_o['ShippedDate'] = df_o['ShippedDate'].astype('datetime64[ns]')
```

```python
# Выведем годы, месяцы, дни и недели в новые столбцы
df_o['Year'] = df_o['OrderDate'].dt.year
df_o['Month'] = df_o['OrderDate'].dt.month
df_o['Day'] = df_o['OrderDate'].dt.day
df_o['Week'] = df_o['OrderDate'].dt.isocalendar().week
df_o['Quarter'] = df_o['OrderDate'].dt.quarter
```

```python
df_o.head()
```

```python
# Группируем по году и месяцу кол-во заказов
df_for_pivoting = df_o.groupby(['Year','Month']).OrderID.count().reset_index()
```

```python
df_for_pivoting
```

```python
# Сводная таблица
df_for_pivoting.pivot_table(index = 'Year', columns = 'Month', values = 'OrderID').reset_index()
```

```python
# Заменим пустые значения
df_for_pivoting.pivot_table(index = 'Year', columns = 'Month', values = 'OrderID', fill_value = 0).reset_index()
```

```python
# Вместо группировки использовать pivot и посчитать кол-во заказов
df_o.pivot_table(index = 'Year', columns = 'Month', values = 'OrderID', fill_value = 0, aggfunc = 'count').reset_index()
```
