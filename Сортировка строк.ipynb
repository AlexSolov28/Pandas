```python
# Сортировка по двум столбцам, но первый столбец отсортируем по возрастанию, а второй по убыванию
df_p.sort_values(['CategoryName','UnitPrice'], ascending = [True, False]).head(30)
```

В параметр ascending передаем список из True и False, где True - сортировка по возрастанию, а False - сортировка по убыванию. Количество элементов в списке параметра ascending = [False, True] равно количеству сортируемых столбцов. Первый элемент списка задает правило сортировки для первого столбца, второй элемент для второго столбца и так далее.

```python
# Если нужно отфильтровать строки, а затем отсортировать оставшиеся строки по возрастанию
df_p[df_p['CategoryName']=='Condiments'].sort_values('UnitPrice')
```
