Типы данных:

* object - текст или символьный тип данных, также часто называют строкой. Также в датафрейме может встретиться и string;
* int64 - целое число;
* float64 - число с плавающей точкой;
* datetime64 - дата и время;
* bool - True False;
* category - текст (ведет себя как число при сортировке).
