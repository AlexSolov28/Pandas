#### Шаблоны выражения или регулярные выражения

Фильтрация строк по какому-либо шаблону

```python
# Все должности, где есть Sales
df_c[df_c['ContactTitle'].str.contains('Sales')]
```

```python
# Слово в нижнем регистре
df_c[df_c['ContactTitle'].str.lower().str.contains('sales')]
```

```python
# Sales только в начале попадут в результат
df_c[df_c['ContactTitle'].str.startswith('Sales')]
```

```python
# Слово в конце попадет в результат
df_c[df_c['ContactTitle'].str.endswith('ager')]
```

```python
df_c[df_c['ContactName'].str.startswith('A')]
```
