Часто бывает, что ключи в таблицах называется по-разному. Например в датафрейме df_c ID клиента - это CustID, а в датафрейме df_o ID клиента - это CustomerID. В таком случае в методе .merge() нужно прописать дополнительные параметры left_on и right_on, где:

left_on - ключ левой таблицы (к которой присоединяем)

right_on - ключ правой таблицы (которую присоединяем)

```python
df_c.merge(df_o, left_on ='CustID', right_on = 'CustomerID')
```

```python
df_o.merge(df_c)
```

```python
df_c.merge(df_o)
```

Атрибут how = 'left'  обозначает, что мы хотим вывести все строки из левой таблицы, независимо от того есть ли соотношение по общему ключу в правой таблице. 

```python
df_c.merge(df_o, how = 'left')
```

Если поменять местами датафреймы, но при этом нужно также вывести клиентов без заказов, тогда указываем тип соединения how = 'right':

```python
df_o.merge(df_c, how = 'right')
```

### Типы соединений (джойны)

* how = 'inner' - этот джойн работает по умолчанию. Его можно не прописывать df_с.merge(df_o);
* how = 'left';
* how = 'right';
* how = 'outer' - часто этот джойн называют full;
* how = 'cross'.
